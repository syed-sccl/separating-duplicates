```python
import pandas as pd
from time import sleep
```

```python
filename = 'Output_Price_Changes_WITH_dups.xlsx'
sheetname = 'Sheet1'
duplicates_df = pd.read_excel(filename,sheetname)
duplicates_df.head()
```

```python
print(duplicates_df.columns.values)
```

```python
duplicates_df['Count'] = duplicates_df.groupby('LAST_UPDATED_ID').cumcount()
duplicates_df.head()
```

```python
duplicates_df.sort_values(by=['Count']).tail()
```

```python
df_iterator = [i for _,i in duplicates_df.groupby('Count')]
```

```python
with pd.ExcelWriter('cleaned_separated_dupes.xlsx', engine='xlsxwriter') as writer:
    for i in range(3):
        df_iterator[i].to_excel(writer, sheet_name ='df_iterator' + str(i), index = False)
```
